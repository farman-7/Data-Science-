```python
Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
```

```python
ANS :- Three ArithmeticError exceptions define are :-
       1. ZeroDivisionError
       2. OverflowError
       3. FloatingPointError
```

```python
#Example of ZeroDivisionError
try :
    10/0
except ZeroDivisionError as e :
    print("I am trying to handle a Zerodivision error \n",e)
```

```python
#Example of OverflowError
import math

try:
    result = math.exp(1000)
except OverflowError:
    result = float('inf')
print("This is OverflowError\n",result)
```

```python
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
```

```python
ANS :- The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence
       or dictionary respectively.
       WE can use LookupError exception class to handle both IndexError and KeyError exception classes.
```

```python
#Example of KeyError
try :
    d = {"key":"farman",1 :[2,3,4,5]}
    print(d["key2"])
except KeyError as e :
    print(e)
```

```python
#Example of IndexError
try :
    l = [2,3,4,5,6]
    print(l[6])
except IndexError as e :
    print(e)
```
